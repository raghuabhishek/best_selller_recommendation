```python
#Eliminating the rows which invlove sellers from China
df=df[~df['businessaddress'].str.contains("CN", na=False)]
```

```python
#Extracting the product count from the string
for index, row in df.iterrows():
    if pd.notnull(row['sellerproductcount']):
        countarr=row['sellerproductcount'].split()
        if 'results' in countarr[1:]:
            df['sellerproductcount'][index]=countarr[countarr.index('results')-1] 
```

```python
#Checking null values
df.isnull().sum()
```

```python
#Extracting number of positive seller ratings and its percentage
import re
for index, row in df.iterrows():
    if pd.notnull(row['sellerratings']):
        percent=row['sellerratings'].split('%')
        if 'last' in row['sellerratings']:
            number_of_ratings=re.findall(r"(\d+) ratings",row['sellerratings'])
        else:
            number_of_ratings=re.findall(r"(\d+) total ratings",row['sellerratings'])
            
        eff_rating=int(percent[0])/100*int(number_of_ratings[0])
        df['sellerratings'][index]=eff_rating
```

```python
#Calculating the moving average for negative rating
df['moving_avg_negative_rating']=0
for index, row in df.iterrows():
    moving_avg=(row['Max % of negative seller ratings - last 30 days']+row['Max % of negative seller ratings - last 90 days']+row['Max % of negative seller ratings - last 12 months'])/3
    df['moving_avg_negative_rating'][index]=moving_avg
```

```python
#Top 10 sellers based on the selection criteria
df["sellerratings"] = pd.to_numeric(df["sellerratings"])
df.nlargest(10, ['sellerratings'])
```
